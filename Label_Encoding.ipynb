### Label Encoding:
* converting the labels into numeric form

```python
# import libraries
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
cancer = pd.read_csv('/content/breast_cancer_data.csv')
cancer.head()
```

```python
# Finding the count of different labels
cancer['diagnosis'].value_counts()
```

```python
# Load the label Encoder function
label_encode = LabelEncoder()
```

```python
labels = label_encode.fit_transform(cancer['diagnosis'])
```

```python
# Appending the labels to the data frame
cancer['target'] = labels
```

```python
cancer.head()
```

0 :- benign

1 :- Malingnant

```python
cancer['target'].value_counts()
```
