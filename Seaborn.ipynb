```python
# import seaborn
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# total bill vs tip dataset

tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# setting theme for plots
sns.set_theme()
```

```python
# relative plot
sns.relplot(data=tips, x = 'total_bill', y='tip', col='time', hue='smoker', style='smoker', size= 'size');
```

```python
# load iris dataset
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
sns.scatterplot(x='sepal_length', y='petal_length', hue='species', data = iris)
```

```python
# load titanic dataset
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
sns.barplot(data = titanic, x = 'sex', y = 'survived', hue='class')
```

```python
ax = sns.countplot(x='class', data=titanic, hue = 'who')
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
ax = sns.countplot(x = 'survived', data = titanic, hue = 'sex')
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
# load house price dataset
from sklearn.datasets import fetch_california_housing
house_boston = fetch_california_housing()

house = pd.DataFrame(house_boston.data, columns= house_boston.feature_names)
house['target'] = house_boston.target
house.head()
```

```python
# Distribution plot

sns.histplot(house['target'])
```

```python
# Correlation
# positive correlation
# negative correlation

correlation = house.corr()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(correlation, cbar=True, square= True, fmt = '.1f', annot= True, annot_kws={'size':8},cmap='Blues')
```
