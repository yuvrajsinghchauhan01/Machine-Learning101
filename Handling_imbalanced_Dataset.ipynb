#### Imbalanced Dataset:
A dataset with an unequal class distribution

```python
# Import libraries

import numpy as np
import pandas as pd
```

```python
# Load the dataset

credit = pd.read_csv('/content/credit_data.csv')

# first 5 rows

credit.head()
```

```python
# Shape
credit.shape
```

```python
# distribution of the two classes

credit['Class'].value_counts()
```

This is highly Imbalanced Dataset

0 --- Legit Transaction

1 --- Fraudulent Transaction

```python
legit = credit[credit['Class']==0]
fraud = credit[credit['Class']==1]
```

```python
legit.shape, fraud.shape
```

```python
# under sampling :- Build a sample dataset containing similar distribution of legit & Fraudulent Transactions

legit_sample = legit.sample(n=70)
```

```python
legit_sample.shape
```

```python
# Concatenate the two data frames
new_dataset = pd.concat([legit_sample, fraud],axis=0)
```
