##### Used for Data visualization

```python
# import
import matplotlib.pyplot as plt
```

```python
# import numpy to get data for our plots
import numpy as np
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

```python
x
```

```python
# Plot data :- sin wave

plt.plot(x,y)
plt.show()
```

```python
# cosine wave
plt.plot(x,z)
plt.show()
```

```python
# Adding title , x-axis, y-axis labels

plt.plot(x,y)
plt.title('sin wave')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
# parabola
x = np.linspace(-10, 10, 20)
y = x ** 2
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x, y, 'r.')
plt.show()
```

```python
x = np.linspace(-5,5,50)
plt.plot(x, np.sin(x),'g-')
plt.plot(x , np.cos(x), 'r--')
plt.show()
```

```python
# Bar plot

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
language = ['English', 'Hindi', 'French', 'Spanish', 'Latin']
people = [7,20,14,5,2]
axe = ax.bar(language, people)

plt.show()
```

```python
# Pie chart

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
language = ['English', 'Hindi', 'French', 'Spanish', 'Latin']
people = [7,20,14,5,2]
ax.pie(people,labels=language, autopct = '%1.1f%%')

plt.show()
```

```python
# Scatter plot

x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
# ax.scatter(x,y, color='r')
ax.scatter(x,z, color='b')
plt.show()
```

```python
# from matplotlib import projections
# 3D scatter plot

fig = plt.figure()
ax = plt.axes(projection='3d')
z = 20 * np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z, c=z, cmap='Blues')
plt.show()
```
